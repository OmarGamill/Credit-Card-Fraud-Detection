```python
import pandas as pd 
```

```python
df=pd.read_csv('train.csv')
df.head(10)
```

```python
df.shape
```

```python
y_train=df['Class'].to_numpy().reshape(-1,1)
x_train=df.drop('Class',axis=1).to_numpy()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score,accuracy_score,classification_report
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_perdic=model.predict(x_train)
accuracy=accuracy_score(y_train,y_perdic)
f1Score=f1_score(y_train, y_perdic)
print(f'accuracy: {accuracy} and f1_score: {f1Score}')
```

```python
report=classification_report(y_train, y_perdic)
print(report)
```

```python
val_data=pd.read_csv('val.csv')
y_val=val_data['Class'].to_numpy().reshape(-1,1)
x_val=val_data.drop('Class', axis=1).to_numpy()
```

```python
y_perdic_val=model.predict(x_val)
report2=classification_report(y_val,y_perdic_val)
print(report2)
```
